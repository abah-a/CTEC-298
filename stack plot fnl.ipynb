```python
##Alpha Bah
##CTEC 298
## Professor Bemley

import matplotlib.pyplot as plt

# Data: Median Home Prices for Maryland, California, Texas, Florida, and New York (2020-2023)
states = ['Maryland', 'California', 'Texas', 'Florida', 'New York']
years = [2020, 2021, 2022, 2023]

# Example median home prices (in dollars) for each state
home_prices = {
    'Maryland': [350000, 370000, 390000, 410000],
    'California': [700000, 750000, 800000, 850000],
    'Texas': [300000, 320000, 340000, 360000],
    'Florida': [275000, 300000, 325000, 350000],
    'New York': [500000, 530000, 560000, 590000]
}

# Stack the data for the plot (stacked data for each year)
maryland = home_prices['Maryland']
california = home_prices['California']
texas = home_prices['Texas']
florida = home_prices['Florida']
new_york = home_prices['New York']

# Create a stack plot
plt.figure(figsize=(12, 8))

plt.stackplot(years, maryland, california, texas, florida, new_york, labels=states, 
              colors=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FFD700'])

# Add titles and labels
plt.title('Stack Plot of Median Home Prices (2020-2023) for Different US States', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Median Home Price ($)', fontsize=12)

# Display legend
plt.legend(title='States', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```
