```python
##Alpha Bah
##CTEC 298
## Professor Bemley

import matplotlib.pyplot as plt

# Data: Median Home Prices for Maryland, California, Texas, Florida, and New York (2020-2023)
states = ['Maryland', 'California', 'Texas', 'Florida', 'New York']

# Example median home prices (average of 2020-2023) for each state
average_home_prices = {
    'Maryland': 370000,  # Average of 350000, 370000, 390000, 410000
    'California': 770000,  # Average of 700000, 750000, 800000, 850000
    'Texas': 330000,  # Average of 300000, 320000, 340000, 360000
    'Florida': 312500,  # Average of 275000, 300000, 325000, 350000
    'New York': 545000  # Average of 500000, 530000, 560000, 590000
}

# Values for the pie chart (total home prices for each state)
prices = list(average_home_prices.values())

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    prices,
    labels=states,
    autopct='%1.1f%%',  # Display the percentage of each state
    startangle=140,  # Start angle for the first segment
    colors=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FFD700']
)

# Add a title
plt.title('Proportion of Median Home Prices in Different US States (2020-2023)', fontsize=14)

# Display the pie chart
plt.tight_layout()
plt.show()
```
