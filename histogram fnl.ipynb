```python
##Alpha Bah
##CTEC 298
## Professor Bemley

import matplotlib.pyplot as plt
import numpy as np

# Data: Median Home Prices for Maryland, California, Texas, Florida, and New York (2020-2023)
states = ['Maryland', 'California', 'Texas', 'Florida', 'New York']
years = [2020, 2021, 2022, 2023]

# Example median home prices (in dollars) for each state
home_prices = {
    'Maryland': [350000, 370000, 390000, 410000],
    'California': [700000, 750000, 800000, 850000],
    'Texas': [300000, 320000, 340000, 360000],
    'Florida': [275000, 300000, 325000, 350000],
    'New York': [500000, 530000, 560000, 590000]
}

# Create a histogram for each state's home prices
plt.figure(figsize=(12, 8))

# Set up the number of bins and the color scheme
bins = np.linspace(250000, 900000, 8)  # Customize the bins for the price range
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FFD700']

# Plot histograms for each state
for i, state in enumerate(states):
    plt.hist(home_prices[state], bins=bins, alpha=0.5, label=state, color=colors[i])

# Add titles and labels
plt.title('Histogram of Median Home Prices (2020-2023) for Different US States', fontsize=16)
plt.xlabel('Median Home Price ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend(title='States', fontsize=10)

# Show the chart
plt.tight_layout()
plt.show()
```
