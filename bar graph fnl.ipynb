```python
##Alpha Bah
##CTEC 298
## Professor Bemley

import matplotlib.pyplot as plt
import numpy as np

# Data: Median Home Prices for Maryland, California, Texas, Florida, and New York (2020-2023)
states = ['Maryland', 'California', 'Texas', 'Florida', 'New York']
years = [2020, 2021, 2022, 2023]

# Example median home prices (in dollars) for each state
home_prices = {
    'Maryland': [350000, 370000, 390000, 410000],
    'California': [700000, 750000, 800000, 850000],
    'Texas': [300000, 320000, 340000, 360000],
    'Florida': [275000, 300000, 325000, 350000],
    'New York': [500000, 530000, 560000, 590000]
}

# Set up bar width and positions
bar_width = 0.15
index = np.arange(len(years))  # Positions for the bars on the x-axis

# Create the bar plot
plt.figure(figsize=(12, 8))

# Plot each state's data as a set of bars
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99', '#FFD700']

for i, state in enumerate(states):
    plt.bar(index + i * bar_width, home_prices[state], bar_width, label=state, color=colors[i])

# Add titles and labels
plt.title('Bar Graph of Median Home Prices (2020-2023) for Different US States', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Median Home Price ($)', fontsize=12)

# Customize x-axis ticks
plt.xticks(index + 2 * bar_width, years)  # Center the labels under the bars
plt.legend(title='States', fontsize=10)

# Show the plot
plt.tight_layout()
plt.show()
```
